```python
# 1. Importar las librerías necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Cargar el archivo train.csv
df = pd.read_csv('train.csv')

# 3. Mostrar las primeras filas del DataFrame
print("Primeras filas del DataFrame:")
display(df.head())

# 4. Mostrar información general del DataFrame
print("\nInformación general del DataFrame:")
df.info()

# 5. Mostrar estadísticas descriptivas
print("\nEstadísticas descriptivas:")
display(df.describe())

# 6. Exploración de datos
# a) Cantidad de estudiantes por programa académico
print("\nCantidad de estudiantes por programa académico:")
estudiantes_por_programa = df['ESTU_PRGM_ACADEMICO'].value_counts()
display(estudiantes_por_programa)

# b) Cantidad de estudiantes por departamento
print("\nCantidad de estudiantes por departamento:")
estudiantes_por_departamento = df['ESTU_PRGM_DEPARTAMENTO'].value_counts()
display(estudiantes_por_departamento)

# c) Distribución de estratos de vivienda
print("\nDistribución de estratos de vivienda:")
estrato_vivienda = df['FAMI_ESTRATOVIVIENDA'].value_counts()
display(estrato_vivienda)

# Visualización de la distribución de estratos de vivienda
sns.countplot(x='FAMI_ESTRATOVIVIENDA', data=df)
plt.title('Distribución de Estratos de Vivienda')
plt.xlabel('Estrato de Vivienda')
plt.ylabel('Cantidad de Estudiantes')
plt.show()

# d) Distribución de la variable FAMI_TIENEINTERNET
print("\nDistribución de FAMI_TIENEINTERNET:")
internet = df['FAMI_TIENEINTERNET'].value_counts()
display(internet)

sns.countplot(x='FAMI_TIENEINTERNET', data=df)
plt.title('Acceso a Internet en el Hogar')
plt.xlabel('Tiene Internet')
plt.ylabel('Cantidad de Estudiantes')
plt.show()

# e) Nivel educativo del padre y la madre
print("\nDistribución de la educación del padre:")
educacion_padre = df['FAMI_EDUCACIONPADRE'].value_counts()
display(educacion_padre)

sns.countplot(y='FAMI_EDUCACIONPADRE', data=df)
plt.title('Nivel Educativo del Padre')
plt.xlabel('Cantidad de Estudiantes')
plt.ylabel('Nivel Educativo')
plt.show()

print("\nDistribución de la educación de la madre:")
educacion_madre = df['FAMI_EDUCACIONMADRE'].value_counts()
display(educacion_madre)

sns.countplot(y='FAMI_EDUCACIONMADRE', data=df)
plt.title('Nivel Educativo de la Madre')
plt.xlabel('Cantidad de Estudiantes')
plt.ylabel('Nivel Educativo')
plt.show()

# f) Distribución de las horas semanales de trabajo de los estudiantes
print("\nDistribución de horas semanales de trabajo de los estudiantes:")
horas_trabajo = df['ESTU_HORASSEMANATRABAJA'].value_counts()
display(horas_trabajo)

sns.countplot(x='ESTU_HORASSEMANATRABAJA', data=df)
plt.title('Horas Semanales de Trabajo')
plt.xlabel('Horas de Trabajo')
plt.ylabel('Cantidad de Estudiantes')
plt.show()

# g) Distribución del rendimiento global de los estudiantes
print("\nDistribución del rendimiento global de los estudiantes:")
rendimiento_global = df['RENDIMIENTO_GLOBAL'].value_counts()
display(rendimiento_global)

sns.countplot(x='RENDIMIENTO_GLOBAL', data=df)
plt.title('Rendimiento Global de los Estudiantes')
plt.xlabel('Rendimiento')
plt.ylabel('Cantidad de Estudiantes')
plt.show()

# 7. Resumen de la exploración
print("\nResumen de la exploración:")
print(f"Total de estudiantes: {df.shape[0]}")
print(f"Total de programas académicos: {df['ESTU_PRGM_ACADEMICO'].nunique()}")
print(f"Total de departamentos: {df['ESTU_PRGM_DEPARTAMENTO'].nunique()}")
print(f"Estratos de vivienda únicos: {df['FAMI_ESTRATOVIVIENDA'].nunique()}")
```
